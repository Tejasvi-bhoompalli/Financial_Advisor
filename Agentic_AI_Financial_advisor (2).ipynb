```python
from PIL import Image
import pytesseract
import re

def calculate_bill_total(image_path):

    # OCR extraction
    img = Image.open(image_path)
    text = pytesseract.image_to_string(img)

    print("===== OCR TEXT =====")
    print(text)

    total = 0.0

    # Process line by line
    for line in text.splitlines():

        # Look for item lines containing "="
        # Example:
        # Chicken Biryani = 350.00
        if "=" in line:

            amounts = re.findall(r'\d+\.\d+|\d+', line)

            if amounts:
                total += float(amounts[-1])

    return total


# Test
total_amount = calculate_bill_total("/content/uploads/swiggy_bill.jpeg")

print("\n===== RESULT =====")
print("Total Amount Spent: ₹", total_amount)
```

```python
def perform_ocr(image_path):

    try:

        from google.cloud import vision
        import io

        client = vision.ImageAnnotatorClient()

        with io.open(image_path,"rb") as image_file:
            content=image_file.read()

        image = vision.Image(content=content)

        response=client.text_detection(image=image)

        return response.text_annotations[0].description

    except Exception:

        from PIL import Image
        import pytesseract

        return pytesseract.image_to_string(
            Image.open(image_path)
        )
```

```python
import re

def extract_expense(text):

    amount_match = re.search(r'₹?\s?(\d+)', text)

    amount = float(amount_match.group(1)) if amount_match else 0

    merchant = "Unknown"

    merchants = [
        "Swiggy",
        "Zomato",
        "Uber",
        "Amazon",
        "Dominos",
        "Starbucks"
    ]

    for m in merchants:
        if m.lower() in text.lower():
            merchant = m

    return merchant, amount
```

```python
extract_expense(
"Paid ₹450 to Swiggy"
)
```

```python
category_map = {

"Swiggy":"Food",
"Zomato":"Food",
"Uber":"Transport",
"Amazon":"Shopping",
"Dominos":"Dining",
"Starbucks":"Dining"

}

def categorize(merchant):

    return category_map.get(
        merchant,
        "Others"
    )
```

```python
def budget_advice(
income,
expenses
):

    savings = income-expenses

    ratio = savings/income

    if ratio<0.2:
        return "Reduce discretionary spending."

    elif ratio<0.3:
        return "Savings acceptable."

    return "Strong savings rate."
```
